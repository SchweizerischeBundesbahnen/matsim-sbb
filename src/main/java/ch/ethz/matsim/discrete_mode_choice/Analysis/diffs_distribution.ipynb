```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 

refrence_scenario = 'stMC'

df = pd.read_csv("allScores.csv", sep=";")
df = df.set_index(df.columns[0])
df
```

```python
#diffs to subtour mode choice
diffs = df - df.loc[refrence_scenario]
diffs
```

```python
#sort values per column
#sorted_values = df.apply(lambda x: x.sort_values(), axis=1)
diffs_sorted = pd.DataFrame(np.sort(diffs.values, axis=1), index=diffs.index, columns=diffs.columns)
diffs_sorted
#drop header TODO
```

```python
fig=plt.figure(figsize=(10, 6), dpi=300)
for index, row in diffs_sorted.iterrows():
    y = np.linspace(0.,1., len(row.dropna()))
    plt.plot(row.dropna(), y, label=index)
plt.legend() 
plt.title('Distribution of Differences Compared to ' + refrence_scenario)
plt.xlabel('Score Difference')
plt.ylabel('CDF')
plt.grid()
```

```python
for index, row in diffs_sorted.iterrows():
    if not index == refrence_scenario:
        fig=plt.figure(figsize=(8, 3), dpi=300)
        plt.scatter(df.loc[refrence_scenario], df.loc[index], marker='x')
        x = np.linspace(min(df.loc[refrence_scenario]), max(df.loc[refrence_scenario]), 1000)
        plt.plot(x, x, linestyle='dotted')
        plt.title(refrence_scenario + ' vs. ' + index)
        plt.xlabel(refrence_scenario)
        plt.ylabel(index)
        plt.grid()
        
```
